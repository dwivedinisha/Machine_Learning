```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# 1. Load dataset
df = pd.read_csv("/placement.csv")

# Display first 5 rows
print("First 5 rows of dataset:")
print(df.head())

# 2. Scatter plot between CGPA and Package
plt.scatter(df['cgpa'], df['package'], color="blue", label="Data points")
plt.xlabel("CGPA")
plt.ylabel("Package (in LPA)")
plt.title("Scatter Plot: CGPA vs Package")
plt.legend()
plt.show()

#  3. Simple Linear Regression
X = df[['cgpa']]   # Independent variable (2D) we have to take input in 2D array as linear regression function onlu take 2D array as input variable
y = df['package']  # Dependent variable

model = LinearRegression()
model.fit(X, y)

# 4. Regression line equation
slope = model.coef_[0]
intercept = model.intercept_
print(f"Regression Line: Package = {slope:.2f} * CGPA + {intercept:.2f}")

# 5. Regression line plot
y_pred = model.predict(X)
plt.scatter(X, y, color="blue", label="Actual data")
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression Line")
plt.xlabel("CGPA")
plt.ylabel("Package (in LPA)")
plt.title("Linear Regression: CGPA vs Package")
plt.legend()
plt.show()

# 6. RÂ² Score & Mean Squared Error
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)
print(f"RÂ² Score: {r2:.3f}")
print(f"Mean Squared Error: {mse:.3f}")

# 7. Prediction for CGPA = 8.5
cgpa_new = np.array([[10]])
predicted_package = model.predict(cgpa_new)[0]
print(f"Predicted Package for CGPA = {cgpa_new}: {predicted_package:.2f} LPA")

# 8. Plot prediction point
plt.scatter(X, y, color="blue", label="Actual data")
plt.plot(X, y_pred, color="red", label="Regression Line")
plt.scatter(8.5, predicted_package, color="green", s=100, label=f"Predicted (8.5, {predicted_package:.2f})")
plt.xlabel("CGPA")
plt.ylabel("Package (in LPA)")
plt.title("Prediction on Regression Line")
plt.legend()
plt.show()
```
